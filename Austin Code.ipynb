```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
```

```python
wd = webdriver.Chrome(options=chrome_options)
```

```python
# Scrapping from website - Yelp
# url_orig=wd.get('https://www.yelp.com/')

# search=wd.find_element(by=By.ID,value='search_location')
# search.send_keys("")
# search.send_keys("Austin,TX")
# search.send_keys(Keys.RETURN)

rest_lis=[]
link=[]

# pages=wd.find_element(by=By.XPATH,value='//*[@id="main-content"]/div/ul/li[22]/div/div[2]/span').text
# no_of_pages=int(pages.split("of ",)[1])
wd.get('https://www.yelp.com/search?find_desc=&find_loc=Austin%2C+TX')
url_temp=wd.current_url

for i in range(0,24):
    url_main=wd.get(url_temp+'&start='+str(i*10))
    try :
        rest_link=wd.find_elements(by = By.CLASS_NAME,value='css-19v1rkv')
        for restaurant in rest_link:
            restaurant_name=restaurant.text
            restaurant_link=restaurant.get_attribute('href')
            rest_lis.append(restaurant_name)
            link.append(restaurant_link)
    except Exception:
            pass
        

    
df_res=pd.DataFrame({'Restaurant Name':rest_lis, 'URL':link})
df_res=df_res.drop_duplicates()
```

```python
import time
no_of_reviews=[]
product_name = []
product_review = []
rest_name=[]
user_rating = []

#looping though all the list of restaurants in first city 

### CHANGE HERE 

for i,row in df_res[:].iterrows():
    
  url=row['URL']
  rest=row['Restaurant Name']
  print('Getting reviews from {}'.format(url))
  page = wd.get(url)
    
  try:
    reviews_cnt = wd.find_element(by = By.XPATH, value = '/html/body/yelp-react-root/div[1]/div[3]/div[1]/div[1]/div/div/div[2]/div[2]/span[2]/a')
  except Exception:
    reviews_cnt=None  
  
  if reviews_cnt is None:
        reviews_cnt1=None
  else:
        reviews_cnt1=reviews_cnt.text
  print(reviews_cnt1)          
  no_of_reviews.append(reviews_cnt1)
  
  time.sleep(5)  
  
  try:
    no_of_pages_per_rest=wd.find_element(by = By.XPATH, value = '//*[@id="reviews"]/section/div[2]/div[5]/div[2]/span').text
  except Exception:
    no_of_pages_per_rest=None
  print(no_of_pages_per_rest)

  if no_of_pages_per_rest !=None:   
          no_of_ppr_int= int(no_of_pages_per_rest.split("of ",)[1])
          print(no_of_ppr_int)

          # Looping through the no. of pages in each restaurant
          min_pag=min(no_of_ppr_int,50)  
          for i in range(0,min_pag):

            url_new=url+'?start='+str(i*10)
            page = wd.get(url_new)
            reviews = wd.find_elements(by = By.CLASS_NAME, value = 'raw__09f24__T4Ezm')
            for i,review in enumerate(reviews):
                if review.get_attribute('lang') =='en': 
                    comment_actual = review.text
                    product_review.append(comment_actual)
                    rest_name.append(rest)
          print(len(product_review))
```

```python
df_comment_final=pd.DataFrame({'Restaurant':rest_name,'Comments':product_review})
```
